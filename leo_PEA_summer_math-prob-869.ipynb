Little kid's homework asks, 

for $$x^2 + 6xy -7y^2 = 20$$

(a) Calculate the distance between the vertices and the distance between the foci

(b) Find parametric equations and sketch the graph using a graphing tool

Let's see what it looks like, first..

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
x, y = np.meshgrid(x, y)

def axes():
    plt.axhline(0, alpha=.5)
    plt.axvline(0, alpha=.5)
    
    
axes()
plt.contour(x, y,x**2 + 6*x*y - 7*y**2, [20], colors='k')
plt.show()
```

What are the asymptotes? It gives a nice trick, for $$x^2 - 6xy +7y^2 = (x-y)(x+7y)$$, the two asymptotes are the two linear terms that takes part in the multiplication, i.e. $x-y = 0$ and $x+7y=0$, let's add that ..

```python
axes()
plt.contour(x, y,x**2 + 6*x*y - 7*y**2, [20], colors='k')
c2= plt.contour(x, y, x+7*y, [0], colors='b')
c3= plt.contour(x, y, x-y, [0], colors='g')
plt.show()
```

We now want the angle bisector, as hinted, in algebraic form, we claim that it is $$\frac{x+7y}{\sqrt{50}} = \frac{x-y}{\sqrt{2}}$$

This stems from the general theorem that the distance of a point $P = (x', y')$ to a line $L: ax + by + c = 0$ is $$d(P,L) = \frac{ax' + by' +c}{\sqrt{a^2 + b^2}}$$

```python
axes()
plt.contour(x, y,x**2 + 6*x*y - 7*y**2, [20], colors='k')
c2= plt.contour(x, y, x+7*y, [0], colors='b')
c3= plt.contour(x, y, x-y, [0], colors='g')
c4 = plt.contour(x, y, (x+7*y)/np.sqrt(50) - (x-y)/np.sqrt(2), [0], colors='r')
plt.show()
```

Now we solve for the vertices as a line intersecting a hyperbola:
$$\begin{align*}
\frac{x+7y}{\sqrt{50}} &= \frac{x-y}{\sqrt{2}}\\
(x+7y)(x-y)&=20
\end{align*}$$
Now, let $A = x + 7y$, and $B = x-y$, then 
$$\begin{align*}
\frac{A}{\sqrt{50}} &= \frac{B}{\sqrt{2}}\\
AB &= 20
\end{align*}$$
This solves to $ A = \pm 10, B = \pm 2$, and then  $$y = (A - B) / 8=\pm1, x= B + y = \pm3$$

Recall the standard  form of a hyperbola:  
$$ \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1$$
thus $a = \sqrt{x^2 + y^2} = \sqrt{10}$.

We know the slope of the asympotes is $b/a$ that is the $\tan(\theta/2)$ where $\tan(\theta) = \frac{m_1 - m_2}{1+ m_1 m_2}$ where $m_1 = 1, m_2=-1/7$ (as per [link](https://www.cuemath.com/geometry/angle-between-two-lines/) ), $\tan(\theta) = (1 + 1/7) / (1-1/7) = 4/3$. Thus $\tan(\theta/2) = \frac{\sin(\theta)}{(1 + \cos(\theta)} = \frac{4/5}{1+3/5} = \frac{1}{2}$ ($\tan(\theta)$ formula as per [wikipedia](https://en.wikipedia.org/wiki/Tangent_half-angle_formula))

Thus $b=\frac{1}{2}a = \frac{\sqrt{10}}{2}$

Thus $c=\sqrt{a^2 + b^2} = \sqrt{10 + 10/4} = \frac{5\sqrt{2}}{2}$

And the distances betweent the vertices is $2a = \boxed{2\sqrt{10}}$ and the distance between the foci is $2c = \boxed{5\sqrt{2}}$

The parametric expression in the standard form is $(a\sec(\alpha), b\tan(\alpha))$, and the rotation angle $\beta$: $\tan(\beta)$ is the slope of $(x+7y = 5x-5y) \equiv (12y = 4x) \equiv (y = 1/3x)$, thus the rotation matrix is $\left(\begin{matrix}c & -s \\ s & c\end{matrix}\right)$ and $c = \cos(\beta) = 1/\sqrt{10}$ and $s = \sin(\beta) = 3/\sqrt{10}$

Thsu the parametric form is
$$ \left(\begin{matrix}1/\sqrt{10} & -3/\sqrt{10}\\3/\sqrt{10} & 1/\sqrt{10}\end{matrix}\right) \left(\begin{matrix}\sqrt{10}\sec(\alpha) \\ \frac{\sqrt{10}}{2}\tan(\alpha)\end{matrix}\right) 
$$

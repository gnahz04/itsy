Little kid's homework asks, what is the vertex of the hyperbola $(x+3y)(x-y)=9$?

Let's see what it looks like, first..

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)

def axes():
    plt.axhline(0, alpha=.5)
    plt.axvline(0, alpha=.5)
    
    
axes()
plt.contour(x, y,(x+3*y)*(x-y), [9], colors='k')
plt.show()
```

And the homework does a lot of hand-holding, it asks: what are the asymptotes? It gives a nice trick, the two asymptotes are the two linear terms that takes part in the multiplication, i.e. $x-3y = 0$ and $x-y=0$, let's add that ..

```python
axes()
plt.contour(x, y,(x+3*y)*(x-y), [9], colors='k')
c2= plt.contour(x, y, x+3*y, [0], colors='b')
c3= plt.contour(x, y, x-y, [0], colors='g')
plt.show()
```

We now want the angle bisector, as hinted, in algebraic form, we claim that it is $$\frac{x+3y}{\sqrt{10}} = \frac{x-y}{\sqrt{2}}$$

This stems from the general theorem that the distance of a point $P = (x', y')$ to a line $L: ax + by + c = 0$ is $$d(P,L) = \frac{ax' + by' +c}{\sqrt{a^2 + b^2}}$$

```python
axes()
plt.contour(x, y,(x+3*y)*(x-y), [9], colors='k')
c2= plt.contour(x, y, x+3*y, [0], colors='b')
c3= plt.contour(x, y, x-y, [0], colors='g')
c4 = plt.contour(x, y, (x+3*y)/np.sqrt(10) - (x-y)/np.sqrt(2), [0], colors='r')
plt.show()
```

Now we solve for the vertices as a line intersecting a hyperbola:
$$\begin{align*}
\frac{x+3y}{\sqrt{10}} &= \frac{x-y}{\sqrt{2}}\\
(x+3y)(x-y)&=9
\end{align*}$$
Now, let $A = x + 3y$, and $B = x-y$, then 
$$\begin{align*}
\frac{A}{\sqrt{10}} &= \frac{B}{\sqrt{2}}\\
AB &= 9
\end{align*}$$
This solves to $ A = \pm 3 \cdot 5^{1/4}, B = \pm 3 \cdot 5^{-1/4}$, and then  $y = (A - B) / 4$ and $x= B + y$

```python
A = 3 * np.power(5, 1.0/4); B= 3*np.power(5, -1.0/4);
y = (A-B)/4.0; x = B + y;
print(f'(x,y) = ({x:.5g}, {y:.5g})')
```
